# **2. Modélisation :**

# a)-Choisir une ou plusieurs fonctions de lien et distributions appropriées pour modéliser le nombre de locations de vélos (variable de comptage)

Pour modéliser le nombre de locations de vélos, qui est une variable de comptage (une variable qui représente un nombre entier positif, comme le nombre de fois qu'un événement se produit), certaines méthodes et fonctions sont particulièrement bien adaptées. Les méthodes classiques pour ce type de données incluent la régression de Poisson et la régression binomiale négative. Ces méthodes utilisent des distributions qui conviennent aux données de comptage et des fonctions qui permettent de bien comprendre la relation entre les facteurs explicatifs (comme la température, l'humidité, etc.) et la variable à prédire (le nombre de locations de vélos).

# b)- Construire et ajuster plusieurs modèles de régression, en utilisant différentes combinaisons de variables explicatives et en tenant compte de la temporalité des données (si pertinent)

. ------Régression linéaire simple-------

La régression linéaire simple est un modèle qui permet de comprendre la relation entre deux variables, c'est-à-dire une variable explicative (ou indépendante) et une variable cible (ou dépendante). Par exemple, nous allons voir comment la température affecte le nombre de locations de vélos.

```python
import statsmodels.api as sm

# Choisir la variable explicative (ici, la température)
X = df_hour[['temp']]  # Température

# Ajouter une constante pour le biais (interception)
X = sm.add_constant(X)

# Variable cible (nombre de locations)
y = df_hour['cnt']

# Ajuster le modèle de régression linéaire
model = sm.OLS(y, X).fit()

# Résultats du modèle
print(model.summary())
```

.------- Régression linéaire multiple--------

La régression linéaire multiple est utilisée pour modéliser la relation entre plusieurs variables explicatives et une variable cible. Par exemple, nous allons voir comment la température, l'humidité et la vitesse du vent affectent le nombre de locations de vélos.

```python
# Choisir plusieurs variables explicatives
X_multiple = df_hour[['temp', 'hum', 'windspeed']]  # Température, humidité et vitesse du vent

# Ajouter une constante
X_multiple = sm.add_constant(X_multiple)

# Variable cible (nombre de locations)
y = df_hour['cnt']

# Ajuster le modèle de régression linéaire multiple
model_multiple = sm.OLS(y, X_multiple).fit()

# Résultats du modèle
print(model_multiple.summary())
```

------Régression de Poisson-----

La régression de Poisson est utilisée lorsque la variable à prédire est une variable de comptage, c'est-à-dire un nombre entier positif (par exemple, le nombre de locations de vélos). Contrairement à la régression linéaire, la régression de Poisson est plus adaptée aux données où la variable cible représente un nombre d'événements, et où la variance est souvent proportionnelle à la moyenne.


```python
# Choisir la variable explicative (Température)
X_poisson = df_hour[['temp']]  # Température

# Ajouter une constante (interception)
X_poisson = sm.add_constant(X_poisson)

# Variable cible (Nombre de locations)
y = df_hour['cnt']

# Ajuster le modèle de régression de Poisson
poisson_model = sm.GLM(y, X_poisson, family=sm.families.Poisson()).fit()

# Résultats du modèle
print(poisson_model.summary())
```

# c)-Comparer les performances des différents modèles et sélectionner le meilleur en utilisant des critères appropriés (AIC, BIC, etc.)

1. Régression linéaire simple;

Variable explicative : Température.

AIC : 227,000.

BIC : 227,000.

R² : 0.164.

Interprétation : Le modèle explique environ 16.4 % de la variance totale des données (R² = 0.164). C'est relativement faible, suggérant que la température seule ne suffit pas à modéliser précisément le nombre de locations. L'AIC est élevé, ce qui suggère que le modèle est loin d'être optimal.

2. Régression linéaire multiple :

Variables explicatives : Température, Humidité, Vitesse du vent.

AIC : 251,000.

BIC : 251,000.

R² : 0.251.

Interprétation : En ajoutant l'humidité et la vitesse du vent, le modèle devient légèrement meilleur, expliquant 25.1 % de la variance des données (R² = 0.251). L'AIC reste assez élevé, mais est un peu inférieur à celui du modèle simple, indiquant une amélioration par rapport au premier modèle.

3. Régression de Poisson :

Variable explicative : Température.

AIC : Non spécifié ici (mais généralement plus faible pour les modèles de Poisson).

BIC : Non spécifié ici (mais généralement plus faible pour les modèles de Poisson).

Pseudo R² : 1.000 (apparemment surévalué dans ce contexte).

Interprétation : La régression de Poisson montre un modèle avec un Pseudo R² très élevé, ce qui pourrait indiquer une sur-adaptation (overfitting) dans ce cas précis. Bien que la température semble être une bonne variable explicative dans le cadre d'un modèle de Poisson, ce résultat peut ne pas être totalement fiable sans une évaluation plus approfondie (comme la comparaison des AIC).

# **Conclusion basée sur AIC et R² :**

AIC : Plus l'AIC est bas, meilleur est le modèle. Le modèle avec la régression multiple a un AIC un peu plus bas, ce qui indique qu'il est légèrement meilleur que le modèle simple. Les modèles de Poisson ne spécifient pas d'AIC ici, mais ils sont souvent plus appropriés pour les variables de comptage.


R² : Le modèle de régression multiple a un meilleur R² (0.251) que la régression simple (0.164), ce qui signifie qu'il explique mieux la variance des données.


Poisson : Le modèle de Poisson présente un Pseudo R² très élevé (1.0), mais cela peut être un signe de sur-ajustement. Il est donc important de vérifier les performances avec de nouveaux jeux de données ou des tests supplémentaires.

# **Recommandation :**

Le modèle de régression linéaire multiple semble être le plus approprié ici, car il améliore significativement l'ajustement (R² de 0.251) par rapport au modèle simple tout en restant simple et sans sur-ajustement apparent. Cependant, si tu cherches à modéliser des variables de comptage comme le nombre de locations de vélos, un modèle de régression de Poisson pourrait être une meilleure option avec un AIC calculé pour confirmer les résultats.

# ***Conclusion sur la qualité du modèle :***

Le modèle de régression linéaire multiple avec la température, l'humidité, et la vitesse du vent explique environ 25,1 % de la variance des données (R² = 0.251). La température et l'humidité sont les facteurs les plus significatifs, avec la température ayant un impact positif et l'humidité ayant un impact négatif. # **3. Prédictions et recommandations :**

# a)-Utiliser le modèle sélectionné pour prédire la demande de vélos à différentes heures de la journée et dans différentes conditions météorologiques

Pour utiliser le modèle de régression linéaire multiple sélectionné afin de prédire la demande de vélos à différentes heures de la journée et sous différentes conditions météorologiques, voici les étapes à suivre :

# Étape 1 :
     Construire un DataFrame avec de nouvelles conditions météorologiques

```python
# Choisir les variables explicatives : Température, Humidité, Vitesse du vent
X_multiple = df_hour[['temp', 'hum', 'windspeed']]

# Ajouter une constante
X_multiple = sm.add_constant(X_multiple)

# Variable cible (nombre de locations)
y = df_hour['cnt']

# Ajuster le modèle de régression linéaire multiple
lin_model_multiple = sm.OLS(y, X_multiple).fit()

# Résultats du modèle
print(lin_model_multiple.summary())
```

# Étape 2 :

 Utiliser le modèle pour faire des prédictions
Une fois que nous avons construit notre DataFrame avec les nouvelles conditions météorologiques, nous utilisons le modèle de régression linéaire multiple que nous avons ajusté pour prédire la demande de vélos (nombre de locations).

```python
# Créer un DataFrame avec de nouvelles conditions météorologiques
new_conditions = pd.DataFrame({
    'temp': [0.5, 0.6, 0.4, 0.7],        # Température pour différentes heures
    'hum': [0.3, 0.4, 0.7, 0.6],         # Humidité
    'windspeed': [0.2, 0.1, 0.3, 0.2]    # Vitesse du vent
})

# Ajouter une constante pour l'interception
new_conditions = sm.add_constant(new_conditions)

# Utiliser le modèle multiple pour prédire la demande
predicted_demand = lin_model_multiple.predict(new_conditions)

# Afficher les prédictions
print("Prédictions du nombre de locations de vélos pour différentes conditions météorologiques :")
print(predicted_demand)
```

# Étape 3 :

 Interprétation des prédictions


Les prédictions du modèle montrent que la température a un effet positif sur la demande de vélos : plus il fait chaud, plus les gens sont enclins à utiliser les vélos, sauf dans des conditions d'humidité élevée. En revanche, une humidité élevée semble avoir un effet négatif, réduisant la demande, comme on le voit dans la troisième condition (temp = 0.4, hum = 0.7).

Ces résultats peuvent nous aider à mieux comprendre comment ajuster la gestion des vélos en fonction des conditions météorologiques. Par exemple, lors de journées plus humides ou plus froides, tu pourrais anticiper une baisse de la demande.
